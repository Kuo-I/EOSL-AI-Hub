```python
import matplotlib.pyplot as plt

# Data Transmission Efficiency
plt.figure(figsize=[15, 4])
plt.subplot(1, 2, 1)
memory = [5, 15, 10, 25, 10, 30, 15, 60]
power = [4,  8,  8, 16, 32, 32, 32, 64]
plt.scatter(power, memory, alpha=0.7, edgecolors='none', label='NVIDIA', color='#2ca02c')
memory = [3, 5, 5, 10, 7, 15, 10, 20]
power = [2, 4, 4, 8, 4, 8, 8, 16]
plt.scatter(power, memory, alpha=0.7, edgecolors='none', label='MediaTek', color='#ff7f0e')
memory = [0.25, 0.5, 1, 1, 1, 2, 4, 8, 4, 8]
power =  [1.8, 3.5,2.5, 4,2.5, 3,4.5, 6, 4, 8]
plt.scatter(power, memory, alpha=0.7, edgecolors='none', label='ARM', color='#17becf')

plt.xlabel('Power (W)'); plt.ylabel('Memory (GB)'); plt.title('Data Transmission Efficiency', loc='left')
plt.grid(True); plt.legend(); plt.gca().spines['top'].set_visible(False); plt.gca().spines['right'].set_visible(False)
plt.xscale('log'); plt.yscale('log'); plt.xticks([10 ** i for i in [0, 0.25, 0.5, 0.75, 1]]); plt.yticks([10 ** i for i in [0, 0.25, 0.5, 0.75, 1]])

# Computational Performance on ResNet50
plt.subplot(1, 2, 2)
fps = [180, 457, 31, 6423]
power = [10, 17.5 ,15, 37.5]
plt.scatter(power, fps, alpha=0.7, edgecolors='none', label='NVIDIA', color='#2ca02c')

fps = [30, 60, 60, 60]
power = [4, 7.5, 11, 15]
plt.scatter(power, fps, alpha=0.7, edgecolors='none', label='ITRI SoC', color='#17becf')

fps = [1.5, 4, 7, 12]
power = [2.65, 3.25, 4.5, 6]
plt.scatter(power, fps, alpha=0.7, edgecolors='none', label='ARM', color='#ff7f0e')

plt.xlabel('Power (W)'); plt.ylabel('Speed (FPS)'); plt.title('Computational Performance', loc='left')
plt.grid(True); plt.gca().spines['top'].set_visible(False); plt.gca().spines['right'].set_visible(False)
plt.xscale('log'); plt.legend(); plt.yscale('log'); plt.xticks([10 ** i for i in [0, 0.25, 0.5, 0.75, 1]]); plt.yticks([10 ** i for i in [0, 0.5, 1, 1.5, 2, 2.5, 3]])
plt.show()
```
